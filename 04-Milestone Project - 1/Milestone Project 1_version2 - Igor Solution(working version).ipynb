```python
def initiate_game():
    global allresponses
    global p1response
    global p2response
    global p
    global pxo
    
    global nrofmoves
    
    global game
    global winning_combinations
    global winning_string
    global p1temp
    global p2temp
    global pl
    global luck

    p = {1:' ',2:' ',3:' ',4:' ',5:' ',6:' ',7:' ',8:' ',9:' '}
    pxo = {1:' ',2:' ',3:' ',4:' ',5:' ',6:' ',7:' ',8:' ',9:' '}
    
    game = True
    
    pl = ''
    p1temp = ''
    p2temp= ''
    luck = ''
    nrofmoves = 0
    
    allresponses = []
    p1response = []
    p2response = []
    p1temp = []
    p2temp = []
    winning_combinations = [['1','2','3'],['4','5','6'],['7','8','9'],\
                            ['1','4','7'],['2','5','8'],['3','6','9'],\
                            ['3','5','7'],['1','5','9']]
    winning_string = ['123','456','789','147','258','369','357','159']
```

```python
def sample_board():
    j = 0
    patterns = {1:'|', 2:'-', 3:' ', 4:'_'}
    for i in range(1,7):
        if i == 2 or i == 4:
            print(2*patterns[4]+ patterns[1]+ 2*patterns[4] + patterns[1] + 2*patterns[4])
        elif i== 6:
            print(2*patterns[3]+ patterns[1]+ 2*patterns[3] + patterns[1] + 2*patterns[3])
        else:
            print(str(j+1) + patterns[3]+patterns[1]+str(j+2)+patterns[3]+patterns[1]+str(j+3))
            j+=3
```

```python
def new_board(p):
    line1 =str(p[1]) + ' ' + '|' + str(p[2]) + ' ' + '|'+str(p[3])+ '\n'
    line2 = 2*'_'+'|'+2*'_'+'|'+2*'_'+'\n'
    line3 = str(p[4]) + ' ' + '|' + str(p[5]) + ' ' + '|'+str(p[6])+'\n'
    line4 = 2*'_'+'|'+2*'_'+'|'+2*'_'+'\n'
    line5 = str(p[7]) + ' ' + '|' + str(p[8]) + ' ' + '|'+str(p[9])+'\n'
    line6 = ' ' + ' ' + '|' + ' ' + ' ' + '|'+' '
    return print (line1 + line2 + line3 + line4 + line5 + line6)
```

```python
def transformxo(p):
    for y in range(1,10):
        if str(p[y]) in p1response:
            pxo[y] = 'X'
        elif str(p[y]) in p2response:
            pxo[y] = 'O'
        elif str(p[y]) == '0':
            pxo[y] = ' '
    return pxo
        
```

```python
def board_xo(pxo):
    linea =str(pxo[1]) + ' ' + '|' + str(pxo[2]) + ' ' + '|'+str(pxo[3])+ '\n'
    lineb = 2*'_'+'|'+2*'_'+'|'+2*'_'+'\n'
    linec = str(pxo[4]) + ' ' + '|' + str(pxo[5]) + ' ' + '|'+str(pxo[6])+'\n'
    lined = 2*'_'+'|'+2*'_'+'|'+2*'_'+'\n'
    linee = str(pxo[7]) + ' ' + '|' + str(pxo[8]) + ' ' + '|'+str(pxo[9])+'\n'
    linef = ' ' + ' ' + '|' + ' ' + ' ' + '|'+' '
    return print (linea + lineb + linec + lined + linee + linef) 
        
```

```python
def comp(m):
    newlist = [i if str(i) in m else 0 for i in range(1,10)]
    return newlist
```

```python
def new_structure(x):
    global p
    for k in range(1,len(comp(x))+1):
        p[k] = x[k-1]
    return p
```

```python
def responses():
    global game
    global p1response
    global p2response
    global allresponses
    global winning_combinations
    global nrofmoves

    global luck
    
    global p1temp
    global p2temp
    
    if nrofmoves == 9:
        game = False
        clear_variables()
        return print('Game Over!')
    
    while True:
        pl1 = input('Player 1, please make your move: ')
        if not pl1.isdigit():
            print('You\'entered a wrong character type')
        elif len(pl1) > 1:
            print('You\'entered too many characters')
        elif str(pl1) in allresponses:
            print('I\'m sorry! You cannot make this move. Try again!')
        else:
            break

    p1response.append(pl1)
    nrofmoves+=1
    p1response.sort()

    p1temp+=pl1
    ''.join(sorted(p1temp))

    allresponses.append(pl1)
    rez = comp(allresponses)

    new_structure(rez)
    transformxo(p)

    clear_output()
    sample_board()
    print('-'*10+'\n')
    
    #new_board(p)
    board_xo(pxo)




    if len(p1response)>=3:

        #print(f'p1response is {p1response} line 20')
        k = 0
        verify(p1response,'Player1')
        while k <= len(p1response)-3:
            verify(sorted(p1response[k:k+3]),'Player 1')
            k+=1
        if len(p1response) >3 and luck =='NO':
            verify_new(p1temp,'Player 1')



    if game:
        if nrofmoves == 9:
            game = False
            clear_variables()
            return ('Game Over!')
        
        while True:
            pl2 = input('Player 2, please make your move: ')

            if not pl2.isdigit():
                print('You\'entered a wrong character type')
            elif len(pl2) > 1:
                print('You\'entered too many characters')
            elif pl2 in allresponses:
                print('I\'m sorry! You cannot make this move. Try again!')
            else:
                break

        p2response.append(pl2)
        nrofmoves+=1
        p2response.sort()

        p2temp+=pl2
        ''.join(sorted(p2temp))

        #sorted(p2response)
        allresponses.append(str(pl2))
        rez = comp(allresponses)

        new_structure(rez)
        #new_structure_pxo(rez)
        transformxo(p)

        clear_output()
        sample_board()
        print('-'*10+'\n')

        #new_board(p)
        board_xo(pxo)


        if len(p2response)>=3:

            #print(f'p2response is {p2response} line 55')
            m = 0
            verify(p2response,'Player2')
            while m <= len(p2response)-3:
                verify(sorted(p2response[m:m+3]),'Player 2')
                m+=1

            if len(p2response) >3 and luck =='NO':
                verify_new(p2temp,'Player 2')

            
```

```python
def verify_new(resp_new,player_new):    
    global game
    global p1response
    global p2response
    global allresponses
    global winning_string
    temp = ''
    counter = 0
    
    resp_new = ''.join(sorted(resp_new))
    
    for i in resp_new:
        match = False
        temp+=i
        if counter >= 3:
            break
            
        counter+=1
        
        for x in winning_string:
            if temp in x[:counter]:
                match=True
        if temp in winning_string:
            game = False
            print(f'{player_new} won!')
            clear_variables()
            break 
        elif match is False and i == resp_new[counter]:
            temp = temp[:-2]
            counter-=2
            
            for j in resp_new[2:]:
                temp+=j
                counter+=1
                
                for x in winning_string:
                    if temp in x[:counter]:
                        match=True
                    elif counter == 3 and temp in winning_string:   
                        break
                    elif counter == 3:
                        temp = temp[:-1]
                        break
                if temp in winning_string:
                    game = False
                    clear_variables()
                    print(f'{player_new} won!')
                    break 
                    
                elif match is False:
                    temp = temp[:-1]
                    counter-=1
            
        elif match is False:
            temp = temp[:-1]
            counter-=1
```

```python
def clear_variables():
    global game
    global p1response
    global p2response
    allresponses = []
    p1response = []
    p2response = []
    luck = ''
```

```python
def verify(resp,player):
    global game
    global p1response
    global p2response
    global allresponses
    global winning_combinations
    global luck
    
    resp.sort()
    #print(f'verify function resp is {resp}')
    if resp in winning_combinations:
        game = False
        print(f'{player} won!')
        clear_variables()
    else:
        luck = 'NO'
```

```python
def play_the_game():
    sample_board()
    initiate_game()
    while game:
        responses()
    clear_variables()
    return ('Game Over!')
```
