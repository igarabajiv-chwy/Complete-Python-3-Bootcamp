```python
import random
class Cards:
    
    def __init__(self,face,color):

        self.play_deck = []
        self.deck = {}

        for i in face:
            for j in color:
                if i == 'ACE':
                    self.play_deck.append(str(i) + ' '+ str(j))
                    continue
                elif type(i) is int:
                    self.deck[str(i) + ' '+ str(j)] = i
                    self.play_deck.append(str(i) + ' '+ str(j))
                else:
                    self.deck[str(i) + ' ' + str(j)] = 10
                    self.play_deck.append(str(i) + ' '+ str(j))

    def get_deck(self,crd):
        return self.deck[crd]

    #def get_play_deck(self):
        #return self.play_deck
    
    def shuffle(self):
        return(random.shuffle(self.play_deck))
```

```python
class hand:

    
    def __init__(self,player):
        self.player = (player.lower()).capitalize()
        self.suites = []
        self.values = 0
        self.amount = 0
    
    def add_suites(self,card):
        self.suites.append(card)
    
    def get_suites(self):
        return self.suites

    def add_hand(self, values):
        self.values += int(values)
    
    def get_values(self):
        return self.values
   
    #initial $100
    def set_bank(self,amount):
        self.amount += int(amount)
        
    def get_bank(self):
        return self.amount
    
    def player_withdraw(self,amount):
        self.amount -= int(amount)
    
    def player_deposit(self,amount):
        self.amount += int(amount)
```

```python
def mix(playcards):
    for i in random.sample(playcards,52):
        yield i
            
```

```python
#winning bet
global game_bet
game_bet = 0
def winning_bet(bet):
    global game_bet
    game_bet += int(bet)
```

```python
def verify_bet_amount(value, player):
    if not value.isdigit():
        print('You entered a wrong character type')
    elif int(value) > int(player.get_bank()):
        print('Insufficient funds!')
    else:
        player.player_withdraw(value)
        winning_bet(value)
        return 'Bet accepted!'
```
